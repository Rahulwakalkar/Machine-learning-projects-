```python
import pandas as pd
from sklearn.model_selection import LeavePOut
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier
from statistics import mean, stdev
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import KFold
from sklearn.model_selection import LeaveOneOut,cross_val_score
from sklearn.model_selection import RepeatedKFold
```

```python
df=pd.read_csv("D:\\PG-DBDA\\Advanced stastical Analysis\\diabetes.csv")
```

```python
df.info()
```

```python
X = df.drop(['Outcome'], axis=1)

y = df['Outcome']
```

```python

# Feature Scaling for input features.
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(X)
```

```python
#Decision tree classifier using gini index
clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)
```

# StratifiedKFold

```python

skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=1)
lst_accu_stratified = []
  
for train_index, test_index in skf.split(X, y):
    x_train_fold, x_test_fold = x_scaled[train_index], x_scaled[test_index]
    y_train_fold, y_test_fold = y[train_index], y[test_index]
    clf_gini .fit(x_train_fold, y_train_fold)
    lst_accu_stratified.append(clf_gini .score(x_test_fold, y_test_fold))
```

```python

print('List of possible accuracy:', lst_accu_stratified)
print('\nMaximum Accuracy That can be obtained from this model is:',
      max(lst_accu_stratified)*100, '%')
print('\nMinimum Accuracy:',
      min(lst_accu_stratified)*100, '%')
print('\nOverall Accuracy:',
      mean(lst_accu_stratified)*100, '%')
print('\nStandard Deviation is:', stdev(lst_accu_stratified))
```

# KFold

```python

kf = KFold(n_splits=4)

scores = []
  
for train_index, test_index in kf.split(X, y):
    x_train_fold, x_test_fold = x_scaled[train_index], x_scaled[test_index]
    y_train_fold, y_test_fold = y[train_index], y[test_index]
    clf_gini .fit(x_train_fold, y_train_fold)
    scores.append(clf_gini .score(x_test_fold, y_test_fold))
```

```python

print('List of possible accuracy:', scores)
print('\nMaximum Accuracy That can be obtained from this model is:',
      max(scores)*100, '%')
print('\nMinimum Accuracy:',
      min(scores)*100, '%')
print('\nOverall Accuracy:',
      mean(scores)*100, '%')
print('\nStandard Deviation is:', stdev(scores))
```

# LeaveOneOut

```python

loo=LeaveOneOut()

scores0 = []

  
for train_index, test_index in loo.split(X, y):
    x_train_fold, x_test_fold = x_scaled[train_index], x_scaled[test_index]
    y_train_fold, y_test_fold = y[train_index], y[test_index]
    clf_gini .fit(x_train_fold, y_train_fold)
    scores0.append(clf_gini .score(x_test_fold, y_test_fold))
```

```python

print('List of possible accuracy:', scores0)
print('\nMaximum Accuracy That can be obtained from this model is:',
      max(scores0)*100, '%')
print('\nMinimum Accuracy:',
      min(scores0)*100, '%')
print('\nOverall Accuracy:',
      mean(scores0)*100, '%')
print('\nStandard Deviation is:', stdev(scores0))
```

# RepeatedKFold

```python

rkf = RepeatedKFold(n_splits=4, n_repeats=2)
scores1 = []
  
for train_index, test_index in rkf.split(X, y):
    x_train_fold, x_test_fold = x_scaled[train_index], x_scaled[test_index]
    y_train_fold, y_test_fold = y[train_index], y[test_index]
    clf_gini .fit(x_train_fold, y_train_fold)
    scores1.append(clf_gini .score(x_test_fold, y_test_fold))
```

```python

print('List of possible accuracy:', scores1)
print('\nMaximum Accuracy That can be obtained from this model is:',
      max(scores1)*100, '%')
print('\nMinimum Accuracy:',
      min(scores1)*100, '%')
print('\nOverall Accuracy:',
      mean(scores1)*100, '%')
print('\nStandard Deviation is:', stdev(scores1))
```

# LeavePout

```python
print('List of possible accuracy:', scores2)
print('\nMaximum Accuracy That can be obtained from this model is:',
      max(scores2)*100, '%')
print('\nMinimum Accuracy:',
      min(scores2)*100, '%')
print('\nOverall Accuracy:',
      mean(scores2)*100, '%')
print('\nStandard Deviation is:', stdev(scores2))
```
