```python
import pandas as pd
from matplotlib import pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
```

```python
df=pd.read_csv("D:\\PG-DBDA\\PML\\RELIANCE.NS.csv")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['SMA30'] =df['Adj Close'].rolling(30).mean()# moving average avg
```

```python
df
```

```python
# moving average of 30 days using .plot() method
df[['Adj Close', 'SMA30']].plot(label='RELIANCE',
                                  figsize=(16, 8))
```

```python
df['CMA30'] = df['Adj Close'].expanding().mean()# calculating cumulative moving avg
```

```python
df['CMA30']
```

```python
df[['Adj Close', 'CMA30']].plot(label='df',
                                  figsize=(16, 8))
```

```python
df['EWMA30'] = df['Adj Close'].ewm(span=25).mean()
```

```python
df[['Adj Close', 'EWMA30']].plot(label='df',
                                   figsize=(16, 8))
```

```python
import matplotlib.pyplot as plt 
from statsmodels.tsa.stattools import adfuller
```

```python
df.plot()
plt.show()
```

```python
result = adfuller(df.Adj Close) # ADF is less than 1% value. print('ADF Statistic: %f' % result[0])
# p-value < 0.05, hence we reject the null hypothesis that data has unit root# It means data is stationary.# Hence d = 0
print('p-value: %f' % result[1])
for key, value in result[4].items(): 
    print('\t%s: %.3f' % (key, value))
```
